```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 15
a = 13

# N = 21
# a = 2
```

```python
math.gcd(a, N)
```

```python
z = list(range(N))
y = [a ** z0 % N for z0 in z]

plt.plot(z, y)
plt.xlabel('z')
plt.ylabel(f'{a}^z mod {N}')
plt.show()
```

```python
r = z[y[1:].index(1)+1]
r
```

```python
if r % 2 == 0:
    x = (a ** (r/2)) % N
    print(f'x: {x}')
    if ((x + 1) % N) != 0:
        print(math.gcd(int(x)+1, N), math.gcd(int(x) - 1, N))
    else:
        print('x + 1 is 0 (mod N)')
else:
    print('r is odd')
```
